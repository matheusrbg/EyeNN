# Welcome to the Eye classifier! 
#### Upload an eye picture and test our classifier :)


```python
import fastbook
fastbook.setup_book()
```

```python
from fastbook import *
from fastai.vision.widgets import *
from ipywidgets import *
```

```python
learn_inf = load_learner('export.pkl')
```

```python
btn_test_upload = SimpleNamespace(data = ['images/blueEyes.jpeg','images/greenEyes.jpeg','images/brownEyes.jpeg'])
out_test_pl1, out_test_pl2, out_test_pl3 = widgets.Output(),widgets.Output(),widgets.Output()
lbl_test_pred1, lbl_test_pred2, lbl_test_pred3 = widgets.Label(),widgets.Label(),widgets.Label()  
```

```python
btn_run = widgets.Button(description='Classify')
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

lbl = {}
out = {}
for i in range(3):
    im = PILImage.create(btn_test_upload.data[i])
    
    out[i] = widgets.Output()
    out[i].clear_output()
    with out[i]: display(im.to_thumb(128,128))

    pred,pred_idx,probs = learn_inf.predict(im)
    lbl[i] = widgets.Label()
    lbl[i].value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```

```python
preVis = [VBox([widgets.Label(f'Test {o+1}'), 
      out[o], lbl[o]]) for o in range(3)]

carousel(preVis)
```

```python
VBox([widgets.Label('Select your eye!'), 
     btn_upload, btn_run, out_pl, lbl_pred])
```
